```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from utils.plot_keypoints import show_keypoints
```

```python
#fb_data = pd.read_csv('data/imputed_body.csv')
#fb_data = pd.read_csv('data/imputed_full_body.csv')
fb_data = pd.read_csv("data/full_body.csv")
```

```python
fb_data
```

```python
fb_data.describe()
```

```python
amplitude = fb_data.max() - fb_data.min()
desvio_padrao = fb_data.std()
estatisticas = pd.DataFrame({'Amplitude': amplitude, 'Desvio Padrão': desvio_padrao})
colunas_ordenadas_por_amplitude = estatisticas['Amplitude'].sort_values(ascending=False)
```

```python
# Plot das amplitudes
plt.figure(figsize=(10, 6))
colunas_ordenadas_por_amplitude.plot(kind='bar')
plt.title('Amplitude das Colunas')
plt.xlabel('Coluna')
plt.ylabel('Amplitude')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
mean_points_raw = fb_data.mean()
mean_predictions = mean_points_raw.values.reshape(-1, 2)
mean_predictions
```

```python
show_keypoints(mean_predictions, save_path='Results/mean_points.jpg')
```

```python
# Calcular a correlação de Pearson
plt.figure(figsize=(20,15))
plt.title("Correlação de Pearson")
sns.heatmap(fb_data.corr(numeric_only=True, method='pearson'), annot=True, cmap='Blues')
plt.show()
```

```python
# Calcular a correlação de Spearman
plt.figure(figsize=(20,15))
plt.title("Correlação de Spearman")
sns.heatmap(fb_data.corr(numeric_only=True, method='spearman'), annot=True, cmap='Blues')
plt.show()
```

```python
num_cols = 34
num_rows = (num_cols + 1) // 2
fig, axes = plt.subplots(nrows=num_rows, ncols=2, figsize=(15, num_rows * 5))

axes = axes.flatten()

# Iterate through columns and create histograms using seaborn
for i, column in enumerate(fb_data.columns):
    ax = axes[i]
    sns.histplot(data=fb_data, x=column, kde=True, ax=ax)
    ax.set_title(column)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')

# If there are unused subplots, remove them
if num_cols < len(axes):
    for i in range(num_cols, len(axes)):
        fig.delaxes(axes[i])


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.stripplot(data=fb_data, orient='h', palette='Set2', jitter=False)
plt.xlabel('Values')
plt.title('Dot Plot of Columns')
plt.show()
```
